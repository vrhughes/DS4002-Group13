```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing our dataset stored in the output folder
df = pd.read_csv('../output/sentiment_data_classified.csv')
```

```python
df
```

```python
df["sentiment_score"] = pd.to_numeric(df["sentiment_score"], errors="coerce")

# count sentiment categories
category_counts = df["sentiment_category"].value_counts()

# bar chart of sentiment counts
plt.figure(figsize=(10, 5))
plt.bar(category_counts.index, category_counts.values, edgecolor="black")
plt.title("Count of BERT Sentiment Classifications")
plt.ylabel("Count")
plt.xlabel("Sentiment Category")
plt.show()
```

```python
# distribution of sentiment scores (histogram)
plt.figure(figsize=(10, 5))
sns.histplot(df["sentiment_score"], bins=30, kde=True, color="blue")
plt.title("Distribution of Sentiment Scores")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# box plot of distribution scores based on tweet category
plt.figure(figsize=(10, 6))
sns.boxplot(x="sentiment", y="sentiment_score", data=df, palette="coolwarm")
plt.title("Distribution of Sentiment Scores per Category")
plt.xlabel("Tweet Category")
plt.ylabel("Sentiment Score")

# relabeling the x axis so that it reflects the category mapping
plt.xticks(ticks=[0, 1, 2, 3], labels=["Anti", "Neutral", "Pro", "News"])
plt.show()
```

```python
# splitting up the dataframe into the tweet categories
df_anti = df[df['sentiment'] == -1]
df_neutral = df[df['sentiment'] == 0]
df_pro = df[df['sentiment'] == 1]
df_news = df[df['sentiment'] == 2]
```

```python
# distribution of sentiment scores (ANTI)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))  # Create a 1-row, 2-column subplot

# histogram
sns.histplot(df_anti["sentiment_score"], bins=30, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Distribution of Sentiment Scores - ANTI Climate Change")
axes[0].set_xlabel("Sentiment Score")
axes[0].set_ylabel("Frequency")

# histogram
sns.barplot(x=df_anti["sentiment_category"].value_counts().index, 
            y=df_anti["sentiment_category"].value_counts().values, 
            palette="coolwarm", ax=axes[1])
axes[1].set_title("Sentiment Category Counts - ANTI Climate Change")
axes[1].set_xlabel("Sentiment Category")
axes[1].set_ylabel("Count")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# distribution of sentiment scores (PRO)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))  # Create a 1-row, 2-column subplot

# histogram
sns.histplot(df_pro["sentiment_score"], bins=30, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Distribution of Sentiment Scores - PRO Climate Change")
axes[0].set_xlabel("Sentiment Score")
axes[0].set_ylabel("Frequency")

# histogram
sns.barplot(x=df_pro["sentiment_category"].value_counts().index, 
            y=df_pro["sentiment_category"].value_counts().values, 
            palette="coolwarm", ax=axes[1])
axes[1].set_title("Sentiment Category Counts - PRO Climate Change")
axes[1].set_xlabel("Sentiment Category")
axes[1].set_ylabel("Count")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# distribution of sentiment scores (NEUTRAL)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))  # Create a 1-row, 2-column subplot

# histogram
sns.histplot(df_neutral["sentiment_score"], bins=30, kde=True, color="blue", ax=axes[0])
axes[0].set_title("Distribution of Sentiment Scores - NEUTRAL Climate Change")
axes[0].set_xlabel("Sentiment Score")
axes[0].set_ylabel("Frequency")

# histogram
sns.barplot(x=df_neutral["sentiment_category"].value_counts().index, 
            y=df_neutral["sentiment_category"].value_counts().values, 
            palette="coolwarm", ax=axes[1])
axes[1].set_title("Sentiment Category Counts - NEUTRAL Climate Change")
axes[1].set_xlabel("Sentiment Category")
axes[1].set_ylabel("Count")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
