```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "/content/sentiment_data_classified (2).csv"
df = pd.read_csv(file_path)

df["sentiment_score"] = pd.to_numeric(df["sentiment_score"], errors="coerce")

# Count sentiment categories
category_counts = df["sentiment_category"].value_counts()

# x and y values
x_values = category_counts.index
y_values = category_counts.values

# bar chart of sentiment counts
plt.figure(figsize=(10, 5))
plt.bar(x_values, y_values, edgecolor="black")
plt.title("Count of Sentiment Classifications")
plt.ylabel("Count")
plt.xlabel("Sentiment Category")
plt.show()
```

```python
import seaborn as sns

# box plot of distribution scores
plt.figure(figsize=(10, 6))
sns.boxplot(x="sentiment_category", y="sentiment_score", data=df, palette="coolwarm")
plt.title("Distribution of Sentiment Scores per Category")
plt.xlabel("Sentiment Category")
plt.ylabel("Sentiment Score")
plt.xticks(rotation=45)
plt.show()
```

```python
# distribution of sentiment scores (histogram)
plt.figure(figsize=(10, 5))
sns.histplot(df["sentiment_score"], bins=30, kde=True, color="blue")
plt.title("Distribution of Sentiment Scores")
plt.xlabel("Sentiment Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# average sentiment score for each sentiment category
category_avg_sentiment = df.groupby("sentiment_category")["sentiment_score"].mean()

# Display the results
print(category_avg_sentiment)

# plot the bar graph
plt.figure(figsize=(10, 5))
category_avg_sentiment.plot(kind="bar", color="blue")
plt.title("Average Sentiment Score per Category")
plt.xlabel("Sentiment Category")
plt.ylabel("Average Sentiment Score")
```
